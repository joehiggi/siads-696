```
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Define the ticker for the currency pair.
# For example, EUR/USD is "EURUSD=X".
# You can find currency tickers on the Yahoo Finance website.
ticker = "EURUSD=X"

# Set the end date to today and start date to 10 years ago.
end_date = datetime.now()
start_date = end_date - timedelta(days=10*365) # Approximation for 10 years

# Download the historical exchange rate data.
# The 'interval' parameter can be set to '1d', '1wk', or '1mo' for daily, weekly, or monthly data.
try:
    data = yf.download(ticker, start=start_date, end=end_date, interval='1d')

    if not data.empty:
        print(f"Successfully retrieved the last 10 years of {ticker} exchange rates:")
        print(data.head())
        print("...")
        print(data.tail())

        # Save the data to a CSV file.
        # This can be useful for further analysis or record-keeping.
        filename = f'{ticker}_last_10_years.csv'
        data.to_csv(filename)
        print(f"\nData saved to {filename}")
    else:
        print(f"No data retrieved for ticker: {ticker}. Please check the ticker symbol.")

except Exception as e:
    print(f"An error occurred: {e}")

```
